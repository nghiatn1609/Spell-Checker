```python
import re
```

create a function

```python
def read_corpus(filename):
  with open(filename, "r") as file:
    lines = file.readlines()
    words = []
    for line in lines:
      words += re.findall(r'\w+', line.lower())

  return words
```

```python
words = read_corpus("./big.txt")
print(f"There are {len(words)} total words in the corpus")
```
